```python
import pandas as pd
import numpy as np
from random import randint
```

```python
df = pd.read_csv('boardgames_sale.csv')
df.head()
```

## Games for sale

```python
sale = df.drop(['rank', 'bgg_url', 'min_time', 'max_time', 'year', 'geek_rating', 'num_votes', 'mechanic', 'image_url', 'designer', 'weight'], axis = 1)
sale = sale[sale.price.isnull() == False].reset_index(drop = True)
sale.price = round(sale.price * 4.44, 2)
sale.head()
```

## Games to rent

```python
"""Generates list of random integers (from 0 to 10), with advantage of zeros."""
def games_count(n = np.shape(sale)[0]):
    a = [randint(0, 10) for i in range(n)]
    b = [0 if randint(0, len(a)) < 500 else i for i in a]
    return b
```

```python
rent = sale[['game_id', 'price']]
```

```python
price = rent.price.values
rent_price = []
for price in price:
    if 0 <= price < 100:
        p = 2
    elif 100 <= price < 500:
        p = 5
    elif 500 <= price < 1000:
        p = 10
    elif 1000 <= price < 1500:
        p = 15
    else:
        p = 20
    rent_price.append(p)
```

```python
rent2 = rent.copy()
rent2['rent_price'] = rent_price
rent2 = rent2.drop(['price'], axis = 1)
rent2 = rent2.loc[rent2.index.repeat(games_count())].reset_index(drop = True)
rent2.insert(0, 'item_id', np.array([*range(len(rent2))]) + 1)
rent2
```

```python
# sale.to_csv('../database/games_for_sale.csv', index = False)
rent2.to_csv('../database/games_to_rent.csv', index = False)
```
